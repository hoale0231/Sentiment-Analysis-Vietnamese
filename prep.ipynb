```python
import os

neg = []
pos = []
 
for filename in os.listdir('data/train/neg'):
    f = os.path.join('data/train/neg', filename)
    if os.path.isfile(f):
        neg.append(open(f, 'r', encoding="utf-8").read())
        
for filename in os.listdir('data/train/pos'):
    f = os.path.join('data/train/pos', filename)
    if os.path.isfile(f):
        pos.append(open(f, 'r', encoding="utf-8").read())
```

```python
def text_preprocess(document):
    # xÃ³a html code
    document = remove_html(document)
    # chuáº©n hÃ³a unicode
    document = convert_unicode(document)
    # chuáº©n hÃ³a cÃ¡ch gÃµ dáº¥u tiáº¿ng Viá»‡t
    document = chuan_hoa_dau_cau_tieng_viet(document)
    # tÃ¡ch tá»«
    document = word_tokenize(document, format="text")
    # Ä‘Æ°a vá» lower
    document = document.lower()
    # xÃ³a cÃ¡c kÃ½ tá»± khÃ´ng cáº§n thiáº¿t
    document = re.sub(r'[^\s\wÃ¡Ã áº£Ã£áº¡Äƒáº¯áº±áº³áºµáº·Ã¢áº¥áº§áº©áº«áº­Ã©Ã¨áº»áº½áº¹Ãªáº¿á»á»ƒá»…á»‡Ã³Ã²á»Ãµá»Ã´á»‘á»“á»•á»—á»™Æ¡á»›á»á»Ÿá»¡á»£Ã­Ã¬á»‰Ä©á»‹ÃºÃ¹á»§Å©á»¥Æ°á»©á»«á»­á»¯á»±Ã½á»³á»·á»¹á»µÄ‘_]',' ',document) 
    # xÃ³a cÃ¡c tá»« 1 kÃ½ tá»± vÃ  sá»‘
    document = remove_singe_char_and_number(document)
    # xÃ³a khoáº£ng tráº¯ng thá»«a
    document = re.sub(r'\s+', ' ', document).strip()
    return document

for i, doc in enumerate(neg):
    neg[i] = text_preprocess(doc)

for i, doc in enumerate(pos):
    pos[i] = text_preprocess(doc)
```
